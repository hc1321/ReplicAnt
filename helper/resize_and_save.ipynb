# Resize images to specified size and write to new folder
This is should not cause any issues during training, as darknet uses relative image coordinates

```python
import cv2
from os import listdir
from os.path import join
```

```python
# define location of dataset and return all files
dataset_location = "C:/Users/Legos/Documents/PhD/FARTS/BENCHMARK/REAL/data/obj"
target_dir = "C:/Users/Legos/Documents/PhD/FARTS/BENCHMARK/REAL/data/obj_smol"

rescale = 0.5

for filename in listdir(dataset_location):
    f = join(dataset_location, filename)
    # checking if it is a file
    if filename[-3:] == "JPG":
        print(f)
        img = cv2.imread(f)
        height, width = img.shape[0:2]
        print(width,height)
        
        img_resized = cv2.resize(img, (int(width*rescale), int(height*rescale)), interpolation = cv2.INTER_AREA)
        cv2.imwrite(join(target_dir, filename), img_resized)
```
